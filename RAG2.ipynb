```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("monopoly.pdf")
pages = loader.load_and_split()
```

```python
pages[0]
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100,
    length_function=len,
    is_separator_regex=False,
)
```

```python
texts = text_splitter.create_documents([page.page_content for page in pages])

print(texts[0])
```

```python
from langchain.vectorstores import Chroma
from langchain.schema import Document
from langchain.embeddings import HuggingFaceEmbeddings


CHROMA_PATH = "chroma"

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

db = Chroma.from_documents(
    documents=texts,
    embedding=embedding_model,
    persist_directory=CHROMA_PATH
)

db.persist()
print(f"Saved {len(texts)} chunks to {CHROMA_PATH}.")
```

```python
#embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db = Chroma.from_documents(
    documents=texts,
    embedding=embedding_model,
    persist_directory=CHROMA_PATH
)
```

```python
query_text ="Who invented Monopoly?"
results = db.similarity_search_with_score(query_text,k=3)#search about it
```

```python
if len(results)==0 or results[0][1]<0.7:
    print("I don't know")
```

```python
from langchain.prompts import PromptTemplate
template = """
Answer the questions on the basis of the context below .
{context}
If you can't
answer the question ,reply "I don't know".
question:{question}
"""
```

```python
context_text = "\n\n---\n\n".join([doc.page_content for doc,score in results])
prompt = PromptTemplate(template=template, input_variables=["context", "question"])

# Format the prompt with your context and question
final_prompt = prompt.format(context=context_text, question=query_text)
print(final_prompt)
```

```python
retriever = db.as_retriever(search_kwargs={"k": 3})
```

```python
from langchain_google_genai import GoogleGenerativeAI
from langchain.chains import RetrievalQA
```

```python
llm = GoogleGenerativeAI(model="chat-bison-001", temperature=0.4)
qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=retriever)

query_text = input("Enter your query:")
result = qa_chain.run(query_text)

print("\n Question:", query_text)
print(" Answer:", result)
```
